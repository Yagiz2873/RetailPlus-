### Outlier Capping By Using Quantiles

#### I've chosen this technique that uses IQR and quantiles for capping. Because Quantity, Price and Monetary have skewed distributions.
#### If you have datas that have normal distributions you can use mean and std for capping instead. 

```python
# UL = 0.75 and LL= 0.25 gave me the best results.
def handleOutliersByCapping(dataframe,var):
    q3 = dataframe[var].quantile(0.75)  
    q1 = dataframe[var].quantile(0.25)
    print("Q3: ",q3)
    print("Q1",q1)
    IQR = q3 - q1
    print("IQR: ",IQR)
    lower_limit = q1 - (IQR * 1.5)
    print("Lower Limit: ",lower_limit)
    upper_limit = q3 + (IQR * 1.5)
    print("Upper Limit: ",upper_limit)
    dataframe[var] = np.where(dataframe[var]> upper_limit, upper_limit,
                        np.where(dataframe[var]< lower_limit, lower_limit,
                          dataframe[var]))
    
    
```

```python
outlier_cols = ["Quantity","Price","Monetary"]
```

```python
for col in outlier_cols:
    handleOutliersByCapping(df_copy_cleaned_final_trial,col)
```

```python
print(today_date)
```

```python
df_rfm["Customer_ID"] = df_rfm["Customer_ID"].astype(str).str.strip()
```

```python
rfm=df_rfm.groupby('Customer_ID',observed=False).agg(
    last_purchase_date=('LitDate', 'max'),  
    Frequency=('Invoice', 'nunique'), 
    Monetary=('Monetary', 'sum'),  
    first_purchase_date=('LitDate', 'min')  
).reset_index()
```

```python
rfm["Customer_ID"] = rfm["Customer_ID"].astype(str).str.strip()
```

```python
rfm['Recency'] = (today_date - rfm['last_purchase_date']).dt.days
rfm['Tenure'] = (today_date - rfm['first_purchase_date']).dt.days
rfm.drop(['last_purchase_date', 'first_purchase_date'], axis=1, inplace=True)
rfm.head()
```

```python
rfm["Recency_Monthly"] = rfm["Recency"]/30
rfm.head()
```

```python
rfm["RecencyScore"] = pd.qcut(rfm['Recency'], 5, labels=[5, 4, 3, 2, 1])
rfm["FrequencyScore"] = pd.qcut(rfm['Frequency'].rank(method='first'), 5, labels=[1, 2, 3, 4, 5])
rfm["MonetaryScore"] = pd.qcut(rfm['Monetary'], 5, labels=[1, 2, 3, 4, 5])
rfm["TenureScore"] = pd.qcut(rfm['Tenure'], 5, labels=[1, 2, 3, 4, 5])
```

```python
rfm["RFM_SCORE"] = (rfm['RecencyScore'].astype(str) +
                    rfm['FrequencyScore'].astype(str) +
                    rfm['MonetaryScore'].astype(str))
```

```python
seg_map = {
    r'[1-2][1-2]': 'Hibernating',
    r'[1-2][3-4]': 'At_Risk',
    r'[1-2]5': 'Cant_Loose',
    r'3[1-2]': 'About_to_Sleep',
    r'33': 'Need_Attention',
    r'[3-4][4-5]': 'Loyal_Customers',
    r'41': 'Promising',
    r'51': 'New_Customers',
    r'[4-5][2-3]': 'Potential_Loyalists',
    r'5[4-5]': 'Champions'
}
```

```python
rfm['Segment'] = rfm['RecencyScore'].astype(str) + rfm['FrequencyScore'].astype(str)
rfm['Segment'] = rfm['Segment'].replace(seg_map, regex=True)
rfm[["Segment", "Recency", "Frequency", "Monetary"]].groupby("Segment").agg(["mean", "count"])
```
